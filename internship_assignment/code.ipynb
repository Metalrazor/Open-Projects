3. Category : The rows representing the category has been stored in the variable -categories-.
4. College : The rows showing the college name and code stored in -college-.

```python
# Creating new dataset for iterating all the categories in colleges
uni = df.iloc[categories, 1:]
category = []

for k, v in uni.iterrows():
    for m in v:
        if isinstance(m, str)==True and len(m) < 10:
            category.append(m)
```

### 3. Preprocessing

```python
df["D"] = df["D"].astype(str)
```

```python
# List of college codes
code = np.unique(
    [i.split(' - ')[0] for i in df.loc[[i for i in college if i in df.index], "D"]]
).tolist()[:-1]

# List of colleges
college = np.unique(
    [i.split(' - ')[-1] for i in df.loc[[i for i in college if i in df.index], "D"]]
).tolist()[:-1]

# Dictionary
code_dict = {int(k):v for k,v in zip(code, college)}
```

```python
code_dict
```

```python
def find_value(x):
    """Function to find value from dictionary"""
    for k, v in code_dict.items():
        if k == x:
            return v
```

```python
new_df = pd.DataFrame()

# Code of the college
new_df["code"] = [int(i.split(' - ')[0][:4]) for i in df.loc[branch_name, "D"]]

# College name
new_df["college"] = new_df["code"].apply(find_value)

# Branch code
new_df["branch_code"] = [i.split(' - ')[0] for i in df.loc[branch_name, "D"]]

# Branch of engineering
new_df["branch"] = [i.split(' - ')[1] for i in df.loc[branch_name, "D"]]

# Cut-off provided by the college
new_df["GOPENH/GOPENS"] = [float(i[1:-1]) for i in df.iloc[cut_offs, 1]]
```

```python
def nest_rem(l):
    """Function to remove list nestings and convert it into a single list"""
    result = []
    for sublist in l:
        for item in sublist:
            result.append(item)

    return result
```

```python
def get_score(data, x):
    """Function to get the cut-off marks for every category"""
    cut, cat = [], []
    cond = pu_cut['college_code'] == x
    
    new = pu_cut[cond].query('branch == "Civil Engineering"').iloc[:, 1:23].reset_index(drop=True)
    for k, v in new.iterrows():
        if k % 2:
            cat.append([i for i in v[v.notna()]])
        else:
            cut.append([i[1:-1] for i in v[v.notna()]])
            
    cat = nest_rem(cat)
    cut = nest_rem(cut)
    return {k:v for k, v in zip(cat, cut)}
```

```python
def enlist_col(data):
    """Function to return a dictionary of the extracted cut-offs for each college"""
    new = data.query('branch == "Civil Engineering"')['college_code'].unique().tolist()
    
    data_dict = {k:v for k, v in zip(new, np.arange(0, len(new)))}
    
    # Update new values in the data dictionary
    x = {}
    for k, v in data_dict.items():
        v = get_score(data, k)
        x.update({k:v})
    
    # Delete irregular keys from the data and update proper key with values
    del x['6176']['(32.4152865)']
    del x['6176']['(58.5245830)']
    
    x['6176']['TFWS'] = '69.0975718'
    del x['6176']['1']
    x['6176'].update({'GOPENO': '58.5245830', 'LOPENO': '47.2074767'})
    return pd.DataFrame(x)
```
